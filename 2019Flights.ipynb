```python
frames = [JAN, FEB, MAR, ARP, MAY, JUNE, JULY, AUG, SEP, OCT,NOV,DEC ]

year2019 = pd.concat(frames)
```

```python
year2019
```

```python
def clean(df):
    df.drop(['Unnamed: 26'],axis=1, inplace= True)
    filter1 = df['ORIGIN_CITY_NAME'].isin(['New York, NY','Los Angeles, CA'])
    filter2 = df['DEST_CITY_NAME'].isin(['New York, NY','Los Angeles, CA'])
    return df[filter1 &filter2]
```

```python
year = clean(year2019)
year
```

```python
year.to_csv('year_2019.csv')
```
